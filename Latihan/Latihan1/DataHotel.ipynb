# Data Integration

```python
import pandas as pd
import numpy as np
```

```python
## Merging
hotel = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika'],
                     'JenisBintang': ['2', '1', '5', '1','2'], 
                     'Alamat': ['Jakarta Pusat', 'Jakarta Utara', 'Jakarta Pusat', 'Jakrta Barat', 'Jakarta Timur']})

nilai = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika'],
                      'Jumlah Kamar': ['1958', '478', '1165', '326','702'],
                      'Rata-RataInap': ['1', '2', '2.5', '1', '2'],
                      'Jumlah Tamu': ['1154', '501', '678', '267', '654']})

result1 = pd.merge(hotel, nilai, on='nama')
result1
```

```python
## Merging different key nama
hotel = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika','Tune', 'Moon', 'Noor'],
                     'JenisBintang': ['2', '1', '5', '1','2', '2.5', '2', '1'], 
                     'Alamat': ['Jakarta Pusat', 'Jakarta Utara', 'Jakarta Pusat', 'Jakrta Barat', 'Jakarta Timur', 'Jakarta Barat', 'Jakarta Barat', 'Jakarta Selatan']})

nilai = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika'],
                      'Jumlah Kamar': ['1958', '478', '1165', '326','702'],
                      'Rata-RataInap': ['1', '2', '2.5', '1', '2'],
                      'Jumlah Tamu': ['1154', '501', '678', '267', '654']})

result2 = pd.merge(hotel, nilai, on='nama')
result2
```

```python
# Ordering variable/column
result1 = result1[['nama', 'JenisBintang', 'Alamat', 'Jumlah Kamar', 'Jumlah Tamu','Rata-RataInap']]
result1
```

```python
# Multiple key
hotel = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika'],
                      'kelas': ['Melati', 'Mawar', 'Kenanga', 'Mawar','Melati'],
                     'JenisBintang': ['2', '1', '5', '1','2'], 
                     'Alamat': ['Jakarta Pusat', 'Jakarta Utara', 'Jakarta Pusat', 'Jakrta Barat', 'Jakarta Timur']})

nilai = pd.DataFrame({'nama': ['Ivory', 'Neo', 'Ibis', 'Fave','Santika'],
                      'kelas': ['Melati', 'Kenanga', 'Mawar', 'Mawar', 'Kenanga'],
                      'Jumlah Kamar': ['1958', '478', '1165', '326','702'],
                      'Rata-RataInap': ['1', '2', '2.5', '1', '2'],
                      'Jumlah Tamu': ['1154', '501', '678', '267', '654']})


result3 = pd.merge(hotel, nilai, on=['nama', 'kelas'])
result3
```

```python
# Merge methods (default how="inner")
result_left = pd.merge(hotel, nilai, how='left', on=['nama', 'kelas'])
result_left
```

```python
# Merge methods (default how="inner")
result_right = pd.merge(hotel, nilai, how='right', on=['nama', 'kelas'])
result_right
```

```python
result_outer = pd.merge(hotel, nilai, how='outer', on=['nama', 'kelas'])
result_outer
```

```python
result_inner = pd.merge(hotel, nilai, how='inner', on=['nama', 'kelas'])
result_inner
```

```python
## Joining 
hotel = pd.DataFrame({'JenisBintang': ['2', '1', '5'], 
                     'Alamat': ['Jakarta Pusat', 'Jakarta Utara', 'Jakarta Pusat']},
                    index=['Ivory', 'Neo', 'Ibis'])
nilai = pd.DataFrame({'Kelas': ['Mawar', 'Melati', 'Kenanga'],
                      'Jumlah Kamar': ['1958', '478', '1165']},
                    index=['Ivory', 'Ibis', 'Fave'])
result = hotel.join(nilai) #Based Biodata ditambahkan nilai
result
```

```python
result = nilai.join(hotel, how='outer') #Seluruh data masuk
result
```

```python
result = nilai.join(hotel, how='inner') # Hanya data yang lengkap
result
```

```python
#The data alignment here is on the indexes (row labels). 
#This same behavior can be achieved using merge plus additional arguments instructing it to use the indexes
result = pd.merge(hotel, nilai, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(hotel, nilai, left_index=True, right_index=True, how='inner')
result
```

```python
hotel = pd.DataFrame({'JenisBintang': ['2', '1', '5','3'], 
                     'Alamat': ['Jakarta Pusat', 'Jakarta Utara', 'Jakarta Pusat', 'Jakarta Barat'],
                    'key': ['Ivory', 'Neo', 'Ivory', 'Neo']})
nilai = pd.DataFrame({'Kelas': ['Mawar', 'Melati'],
                      'Jumlah Kamar': ['1958', '478']},
                    index=['Ivory', 'Neo'])
result = hotel.join(nilai, on='key')
result
```

```python
## Concatenating
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                'B': ['B0', 'B1', 'B2', 'B3'],
                'C': ['C0', 'C1', 'C2', 'C3'],
                'D': ['D0', 'D1', 'D2', 'D3']})
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                'B': ['B4', 'B5', 'B6', 'B7'],
                'C': ['C4', 'C5', 'C6', 'C7'],
                'D': ['D4', 'D5', 'D6', 'D7']})
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                'B': ['B8', 'B9', 'B10', 'B11'],
                'C': ['C8', 'C9', 'C10', 'C11'],
                'D': ['D8', 'D9', 'D10', 'D11']})

frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

```python
# Add category
result = pd.concat(frames, keys=['K', 'L', 'M'])
result
```

```python
# Default join=outer (join all rows)
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])

result = pd.concat([df1, df4], axis=0) #axis=1 is for concated by y-axis; default axis=0
result
```

```python
# Join = inner (join only rows that exist in both datasets) 
# hanya yang ada di df1 dan df4
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

```python
# Join the exact rows in df1 
#Menggabungkan berdasarkan df1
result = pd.concat([df1, df4], axis=1, join_axes=[df1.index])
result
```

```python
## Concatenating using append
result = df1.append(df2) #default axis=0
result
```

```python
result = df1.append(df4)
result
```

```python
result = df1.append([df2, df3]) #multiple datasets to concatenate
result
```

```python
# For DataFrames which dont have a meaningful index, 
#you may wish to append them and ignore the fact that they may have overlapping indexes.
result = pd.concat([df1, df4], ignore_index=True)
result
```

```python
result = df1.append(df4, ignore_index=True)
result
```

```python
# Appending rows to a DataFrame
s2 = pd.Series(['K0', 'K1', 'K2', 'K3'], index=['A', 'B', 'C', 'D'])
result = df1.append(s2, ignore_index=True)
result
```

```python
dicts = [{'A': 1, 'B': 2, 'C': 3, 'K': 13}, #dictionary
         {'A': 5, 'B': 6, 'C': 7, 'L': 5}]
result = df1.append(dicts, ignore_index=True)
result
```
