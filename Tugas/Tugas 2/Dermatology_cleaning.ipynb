## Imputation Missing Value

```python
np.sum(data.isnull()) #jumlah missing value masing-masing variabel
```

```python
data.isnull().sum() #menghitung missing value dengan cara lain
```

```python
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik
data['band-like infiltrate'] = data['band-like infiltrate'].fillna((data['band-like infiltrate'].mean())) 
data['Oral'] = data['Oral'].fillna((data['Oral'].median()))
data['PNL'] = data['PNL'].fillna((data['PNL'].median()))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Family history'] = LabelEncoder().fit_transform(data['Family history'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Erythema'] = StandardScaler().fit(data[['Erythema']]).transform(data[['Erythema']])
data['Scaling'] = StandardScaler().fit(data[['Scaling']]).transform(data[['Scaling']])
data['Definite borders'] = StandardScaler().fit(data[['Definite borders']]).transform(data[['Definite borders']])
data['Itching'] = StandardScaler().fit(data[['Itching']]).transform(data[['Itching']])
data['Kobner Phenomenon'] = StandardScaler().fit(data[['Kobner Phenomenon']]).transform(data[['Kobner Phenomenon']])
data['polygonal'] = StandardScaler().fit(data[['polygonal']]).transform(data[['polygonal']])
data['follicular'] = StandardScaler().fit(data[['follicular']]).transform(data[['follicular']])
data['Oral'] = StandardScaler().fit(data[['Oral']]).transform(data[['Oral']])
data['Knee'] = StandardScaler().fit(data[['Knee']]).transform(data[['Knee']])
data['Scalp'] = StandardScaler().fit(data[['Scalp']]).transform(data[['Scalp']])
data['Melanin'] = StandardScaler().fit(data[['Melanin']]).transform(data[['Melanin']])
data['Eosinophils'] = StandardScaler().fit(data[['Eosinophils']]).transform(data[['Eosinophils']])
data['PNL'] = StandardScaler().fit(data[['PNL']]).transform(data[['PNL']])
data['fibrosis'] = StandardScaler().fit(data[['fibrosis']]).transform(data[['fibrosis']])
data['Exocytosis'] = StandardScaler().fit(data[['Exocytosis']]).transform(data[['Exocytosis']])
data['acanthosis'] = StandardScaler().fit(data[['acanthosis']]).transform(data[['acanthosis']])
data['hyperkeratosis'] = StandardScaler().fit(data[['hyperkeratosis']]).transform(data[['hyperkeratosis']])
data['inflammatory'] = StandardScaler().fit(data[['inflammatory']]).transform(data[['inflammatory']])
data['band-like infiltrate'] = StandardScaler().fit(data[['band-like infiltrate']]).transform(data[['band-like infiltrate']])
data['age'] = StandardScaler().fit(data[['age']]).transform(data[['age']])
data.head()
```
