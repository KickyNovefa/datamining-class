# Data Integration

```python
### Data integration
import pandas as pd
import numpy as np
```

```python
#load data
Data = pd.read_csv ('DataSungai.csv','sep',';')
Data1 = pd.read_csv('DataSungai1.csv','sep',';')
Data
```

```python
Data1
```

```python
# merge data based on Stasion
Datas = pd.merge(Data, Data1, on='Station')
Datas
```

```python
#multiple key
Datass = pd.merge(Data, Data1, on = ['Station', 'Ph'])
Datass
```

```python
# unsert new colomn
Datass['Kandungan Kimia'] = ((Datass['Nitrogen'])+(Datass['Fosfor']))
Datass
```

```python
# delete colomn
Datass = Datass.drop(['Derajat_Keasaman'], axis=1)
Datass
```

```python
# Drop baris ke-9
Datass1 = Datass.drop([10])
Datass1
```

```python
# Merge methods (default how="inner")
result_left = pd.merge(Data, Data1, how='left', on=['Station', 'Ph'])
result_left
```

```python
# Merge methods (default how="inner")
result_right = pd.merge(Data, Data1, how='right', on=['Station', 'Ph'])
result_right
```

```python
result_outer = pd.merge(Data, Data1, how='outer', on=['Station', 'Ph'])
result_outer
```

```python
result_inner = pd.merge(Data, Data1, how='inner', on=['Station', 'Ph'])
result_inner
```
